```python
# Flight Data Analysis
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset
df = pd.read_excel('flight_data.xlsx')
# Display first few rows
print("Dataset Preview:")
print(df.head())
# a. Summary statistics for departure delays
print("\nSummary Statistics for Departure Delays:")
print(df['departure_delay'].describe())
# b. Average delay for each airline
print("\nAverage Delay for Each Airline:")
avg_delay = df.groupby('airline')['departure_delay'].mean()
print(avg_delay)
# c. Most common departure and arrival destinations
print("\nMost Common Departure Airport:")
print(df['origin'].value_counts().idxmax())
print("\nMost Common Arrival Airport:")
print(df['destination'].value_counts().idxmax())
# d. Box plot for distribution of departure delays
plt.figure()
plt.boxplot(df['departure_delay'])
plt.title("Distribution of Departure Delays")
plt.ylabel("Delay (minutes)")
plt.show()
```
