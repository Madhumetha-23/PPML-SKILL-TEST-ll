```python
import pandas as pd
df = pd.read_excel("'student_scores.csv'.xlsx")
print("Dataset:\n", df)
# ---------------------------------------------------
# (a) Summary statistics for each subject
# ---------------------------------------------------
print("\nSummary Statistics:")
print(df[["MATHS", "PHYSICS", "CHEMISTRY"]].describe())
# ---------------------------------------------------
# (b) Average score for each student
# ---------------------------------------------------
df["Average"] = df[["MATHS", "PHYSICS", "CHEMISTRY"]].mean(axis=1)
print("\nAverage Score of Each Student:")
print(df[["NAME", "Average"]])
# ---------------------------------------------------
# (c) Students scoring below 60 in more than 2 subjects
# ---------------------------------------------------
low_scores = (df[["MATHS", "PHYSICS", "CHEMISTRY"]] < 60).sum(axis=1)
result = df[low_scores > 2]
print("\nStudents scoring below 60 in more than 2 subjects:")
print(result[["NAME", "MATHS", "PHYSICS", "CHEMISTRY"]])
```
