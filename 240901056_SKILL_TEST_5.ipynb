```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# ---------------------------------------------------
# Step 1: Create Time Series Data (1 Year)
# ---------------------------------------------------
# Generate date range for 1 year
dates = pd.date_range(start="2023-01-01", periods=365, freq="D")
# Generate random temperature data (in °C)
np.random.seed(0)
temperature = np.random.normal(loc=25, scale=5, size=365)
# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Temperature": temperature
})
df.set_index("Date", inplace=True)
print("Dataset Preview:\n", df.head())
# ---------------------------------------------------
# Step 2: Basic Analysis
# ---------------------------------------------------
print("\nSummary Statistics:")
print(df.describe())
# Monthly average temperature
monthly_avg = df.resample("M").mean()
print("\nMonthly Average Temperature:")
print(monthly_avg)
# ---------------------------------------------------
# Step 3: Visualization
# ---------------------------------------------------
# Plot daily temperature
plt.figure(figsize=(10,5))
plt.plot(df.index, df["Temperature"], label="Daily Temp", color="blue")
# Plot monthly average
plt.plot(monthly_avg.index, monthly_avg["Temperature"],
         label="Monthly Avg", color="red", linewidth=2)
plt.title("Daily Temperature Time Series")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid()
plt.show()
```
